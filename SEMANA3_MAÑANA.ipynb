necsito que 5 estudiantes me den sus edaddes y sacar la media y el pomedio y despues de essos resultdos; enviar en barras los resultados

```python
import numpy as np
import matplotlib.pyplot as plt

edad1 = int(input("Ingrese la edad del primer estudiante: "))
edad2 = int(input("Ingrese la edad del segundo estudiante: "))
edad3 = int(input("Ingrese la edad del tercer estudiante: "))
edad4 = int(input("Ingrese la edad del cuarto estudiante: "))
edad5 = int(input("Ingrese la edad del quinto estudiante: "))

edades = np.array([edad1, edad2, edad3, edad4, edad5])
media = np.mean(edades)
promedio = np.average(edades)
print("El promedio de las edades es:", promedio)
print("La media de las edades es:", media)

estudiantes = ['1', '2', '3', '4', '5']
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.bar(estudiantes, edades)
plt.title("Edades de los Estudiantes")
plt.xlabel("Estudiantes")
plt.ylabel("Edad")
plt.ylim(0, max(edades) + 5)

plt.subplot(1, 2, 2)
plt.bar(['Media', 'Promedio'], [media, promedio])
plt.title("Media y Promedio de las Edades")
plt.ylabel("Edad")
plt.ylim(0, max(edades) + 5)

plt.tight_layout()
plt.show()


```

TAREA:DEFINIR CUANTOS ESTUDIANTES SON (SACAR DOS GRAFICOS CON EL EJERCICIO ANTERIOR TOMANDO EN CUENTA LA MEDIA Y LA DESVIACION ESTANDAR PERO DE SUS PESOS) , DE LA CANTIDAD DE ESTUDIANTES EL PORCENTAJE DE VARONES Y MUJERES(Y OTRO GRAFICO ); AMBOS EN DIAGRAMA DE PASTELES

```python
import numpy as np
import matplotlib.pyplot as plt
peso1 = int(input("Ingrese el peso del primer estudiante: "))
peso2 = int(input("Ingrese el peso del segundo estudiante: "))
peso3 = int(input("Ingrese el peso del tercer estudiante: "))
peso4 = int(input("Ingrese el peso del cuarto estudiante: "))
peso5 = int(input("Ingrese el peso del quinto estudiante: "))

peso = np.array([peso1, peso2, peso3,peso4,peso5])
media = np.mean(peso)
desviacion_peso = np.std(peso)
print("La media de las edades es: ", media)
print("El promedio de las edades es: ", desviacion_peso ,'\n')

fig, ax = plt.subplots()
ax.pie([media, desviacion_peso], labels=['Media', 'Desviacion Estandar'])
plt.show()
```
